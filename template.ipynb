```python
reload(cmplt);
# Configuration paramters for the figure

# Panel structure. Each entry different '(a)' is a different panel. 
# Repeated entries becomes merged panels
mosaic = [
    ['(a)', '(b)',],
    ['(a)', '(c)',]
    ]

# The properties of each panel can be configured individually

# Panel (a) controls
a_kwargs = dict(
    ylabel=r'data $1$',
    xlabel=r'some x-axis',
    xticks = (0,1),
    yticks = (0, 1),
)

# Panel (b) controls
b_kwargs = dict(
    ylabel=r'data $2$',
    xlabel='',
    xticks = (),
    yticks = (-1, 1),
)

# Panel (c) controls
c_kwargs = dict(
    ylabel=r'data $3$',
    xlabel=r'another x-axis',
    xticks = (0,),
    yticks = (0, 1)
)

# There are also global properties for the figure

kwargs = dict(
    figsize=(6.8,5.5),
    dpi=500
)

# Relating panels to data and data types
fMosaic = { # ... a structure relating the mosaic to the files and the components that should be plotted
    '(a)': dict(type='templateDATA', data=DATAList[0], panel_kwargs=a_kwargs),
    '(b)': dict(type='templateDATA', data=DATAList[1], panel_kwargs=b_kwargs),
    '(c)': dict(type='templateDATA', data=DATAList[2], panel_kwargs=c_kwargs),

}
```

```python
reload(cmplt);
fig, axes = cmplt.fg.makeFigure(mosaic, fMosaic, **kwargs)
```
